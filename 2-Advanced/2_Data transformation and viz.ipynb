# Analyze Data Science Jobs

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
data = load_dataset('lukebarousse/data_jobs')
df = data['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country']=='United States']
df_US
```

```python
df_US['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc = 'size')
```

```python
df_US_pivot
```

```python
df_US_pivot.reset_index(inplace=True)
```

```python
df_US_pivot
```

```python
df_US_pivot['month_number'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
```

```python
df_US_pivot.sort_values('month_number', inplace=True)
df_US_pivot.set_index('job_posted_month',inplace= True)
df_US_pivot.drop(columns='month_number',inplace=True)
```

```python
df_US_pivot.plot(kind='line')
```

```python
TOP_3 = df_US['job_title_short'].value_counts().head(3)
TOP_3 = TOP_3.index.to_list()
```

```python
df_US_pivot[TOP_3].plot(kind='line')
plt.title('Salary by Job and Month')
plt.xlabel('Month')
plt.ylabel('Salary')
```

# Explode() function

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
data = load_dataset('lukebarousse/data_jobs')
df = data['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list : ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_exploded = df.explode('job_skills')
skills_count = df_exploded.groupby(['job_skills','job_title']).size()
df_skills_count = skills_count.reset_index(name='skills_count')
df_skills_count = df_skills_count.sort_values('skills_count', ascending=False)
```

```python
job_title = "Data Analyst"
top_skills = 10

df_skills_final = df_skills_count[df_skills_count['job_title']==job_title].head(top_skills)
```

```python
df_skills_final.plot(kind='barh',x='job_title',y='skill_count')
```

```python
df_exploded['job_skills'].value_counts().head(10).plot(kind='barh')
plt.gca().invert_yaxis()
plt.title(f'Top {top_skills} skills needed by the {job_title}s')
plt.ylabel('Job Posting Count')
plt.legend().set_visible(False)
```

# Top 5 skills for Data Analyst for this yeat

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
data = load_dataset('lukebarousse/data_jobs')
df = data['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list : ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_DA = df[df['job_title_short']== "Data Analyst"].copy()
```

```python
df_DA['job_posted_month_number'] = df_DA['job_posted_date'].dt.month
```

```python
df_DA_explode = df_DA.explode('job_skills')
df_DA_explode
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_number',columns='job_skills',aggfunc='size', fill_value=0)
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
```

```python
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
df_DA_pivot = df_DA_pivot.drop('Total')
```

```python
df_DA_pivot.iloc[:,:5].plot(kind='line')
plt.title('Top 5 skills for Data Analysis per Month')
plt.ylabel('Count')
plt.xlabel('')
```
